### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

tienda1.head()
```

#1. Análisis de facturación



```python
sumas_totales = []
for i, tienda in enumerate(tiendas):
    suma_precios = tienda["Precio"].sum()
    print(f"La suma total de los precios en {nombres_tiendas[i]} es: {suma_precios}")
    sumas_totales.append(suma_precios)
```

```python
plt.figure(figsize=(10, 6))
colores = ['royalblue', 'firebrick', 'limegreen', 'gold']
plt.bar(nombres_tiendas, sumas_totales, color=colores)
plt.xlabel("Tienda")
plt.ylabel("Suma Total de Precios")
plt.title("Suma Total de Precios por Tienda")
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
for i, tienda in enumerate(tiendas):
  print(f"\nConteo de ventas por categoría en {nombres_tiendas[i]}:")
  conteo_categorias = tienda["Categoría del Producto"].value_counts()
  print(conteo_categorias)
```

```python
colores_categorias = plt.cm.get_cmap('viridis', max(tienda["Categoría del Producto"].nunique() for tienda in tiendas))

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
axes = axes.flatten()

for i, tienda in enumerate(tiendas):
    conteo_categorias = tienda["Categoría del Producto"].value_counts()
    categorias = conteo_categorias.index
    ventas = conteo_categorias.values

    ax = axes[i]
    bars = ax.bar(categorias, ventas, color=[colores_categorias(j) for j in range(len(categorias))])
    ax.set_xlabel("Categoría del Producto")
    ax.set_ylabel("Número de Ventas")
    ax.set_title(f"Ventas por Categoría en {nombres_tiendas[i]}")
    ax.set_xticklabels(categorias, rotation=45, ha='right')
    ax.grid(axis='y', linestyle='--')

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedios_calificacion = []
for i, tienda in enumerate(tiendas):
    promedio_calificacion = tienda["Calificación"].mean()
    print(f"La calificación promedio de la {nombres_tiendas[i]} es: {promedio_calificacion:.2f}")
    promedios_calificacion.append(promedio_calificacion)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(nombres_tiendas, promedios_calificacion, s=100, color=['lightcoral', 'lightseagreen', 'lightskyblue', 'lightsalmon'])
plt.xlabel("Tienda")
plt.ylabel("Calificación Promedio")
plt.title("Calificación Promedio por Tienda")
min_calificacion = min(promedios_calificacion)
max_calificacion = max(promedios_calificacion)
rango_calificacion = max_calificacion - min_calificacion
margen = rango_calificacion * 0.1  # Añadir un pequeño margen (10% del rango)
plt.ylim(min_calificacion - margen, max_calificacion + margen)

plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
for i, tienda in enumerate(tiendas):
    print(f"\nProductos más y menos vendidas en {nombres_tiendas[i]}:")

    # Conteo de la frecuencia de cada producto
    conteo_producto = tienda["Producto"].value_counts()

    # Top 3 de producto más vendidas
    top_3_mas_vendidos = conteo_producto.head(3)
    print("\nTop 3 productos más vendidas:")
    print(top_3_mas_vendidos)

    # Top 3 de producto menos vendidas
    top_3_menos_vendidos = conteo_producto.tail(3)
    print("\nTop 3 productos menos vendidas:")
    print(top_3_menos_vendidos)
```

```python
colores_mas_vendidos = ['blue', 'red', 'green']
colores_menos_vendidos = ['royalblue', 'firebrick', 'limegreen']

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(16, 18))
axes = axes.flatten()

for i, tienda in enumerate(tiendas):
    conteo_producto = tienda["Producto"].value_counts()

    # Top 3 más vendidos
    top_3_mas_vendidos = conteo_producto.head(3)
    productos_mas_vendidos = top_3_mas_vendidos.index
    ventas_mas_vendidos = top_3_mas_vendidos.values

    ax_mas_vendidos = axes[i * 2]
    ax_mas_vendidos.barh(productos_mas_vendidos, ventas_mas_vendidos, color=colores_mas_vendidos)
    ax_mas_vendidos.set_xlabel("Número de Ventas")
    ax_mas_vendidos.set_ylabel("Producto")
    ax_mas_vendidos.set_title(f"Top 3 Más Vendidos en {nombres_tiendas[i]}")
    ax_mas_vendidos.invert_yaxis()
    ax_mas_vendidos.grid(axis='x', linestyle='--')

    # Top 3 menos vendidos
    top_3_menos_vendidos = conteo_producto.tail(3)
    productos_menos_vendidos = top_3_menos_vendidos.index
    ventas_menos_vendidos = top_3_menos_vendidos.values

    ax_menos_vendidos = axes[i * 2 + 1]
    ax_menos_vendidos.barh(productos_menos_vendidos, ventas_menos_vendidos, color=colores_menos_vendidos)
    ax_menos_vendidos.set_xlabel("Número de Ventas")
    ax_menos_vendidos.set_ylabel("Producto")
    ax_menos_vendidos.set_title(f"Top 3 Menos Vendidos en {nombres_tiendas[i]}")
    ax_menos_vendidos.invert_yaxis() # Para que el menos vendido (con menor índice original) esté arriba
    ax_menos_vendidos.grid(axis='x', linestyle='--')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
promedios_costo_envio = []
for i, tienda in enumerate(tiendas):
    promedio_costo_envio = tienda["Costo de envío"].mean()
    print(f"El costo de envío promedio en {nombres_tiendas[i]} es: {promedio_costo_envio:.2f}")
    promedios_costo_envio.append(promedio_costo_envio)
```

```python
total_costo_envio = []
for i, tienda in enumerate(tiendas):
    total_costo_envio = tienda["Costo de envío"].sum()
    print(f"El costo de envío total en {nombres_tiendas[i]} es: {total_costo_envio:.2f}")
```

```python
plt.figure(figsize=(8, 6))
plt.bar(nombres_tiendas, promedios_costo_envio, color=['blue', 'red', 'green', 'gold'])
plt.xlabel("Tienda")
plt.ylabel("Costo de Envío Promedio")
plt.title("Costo de Envío Promedio por Tienda")
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

Lo que podemos observar es que **la tienda 1 es la que tiene una mayor facturacion**, mientras que **la tienda 4 es la que tiene la menor facturacion de las 4 tiendas**.

Otra de las partes mas importantes a tener en cuenta en este analisis al menos desde mi punto de vista es la cantidad de dinero que se gasta en costes de envio de cada producto vendido
Observando el total de costos de envio de cada una de las tiendas, observamos que la **tienda 1 es la que mas gasta en costos de envio**, **mientras que la tienda 4 es la que menos**, sin embargo esto es solo un reflejo de que la tienda 1 una realiza mas transacciones que la tienda 4, como podemos observar en las cantidades de productos vendidos tanto por categoría como por producto.

Si tomamos aparte las calificaciones de las tiendas, observamos que aunque la tienda 1 es la peor calificada, pero la tienda 4 es la segunda peor calificada, por lo que no creo que deberia tener tanta relevancia a la hora de determinar cual es la mejor opcion para vender.

A pesar de que la tienda 4 genera menos gastos de envio, la cantidad de dinero que produce es mucho menor que la de las otras 3 tiendas, por lo tanto considero que **la tienda 4 es la mejor opcion para vender**
